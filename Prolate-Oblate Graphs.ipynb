Radius of Curvature Distribution in Prolate Spheroid
===

Mass function:

$\huge\displaystyle f_C(\bar{R}) = \frac{\sqrt{\frac{C^{4/3}\cdot\bar{R}^{2/3}-1}{C^4-C}}}{\frac{3C^2}{4} - \frac{3}{8C} - \frac{3\cdot\log\left(C^{3/2}+\sqrt{C^3-1}\right)}{8C^2\sqrt{C^4-C}}}$

Cumulative function:

$\large F_C(\bar{R}) = \frac{\displaystyle\sqrt{\frac{C^{4/3}\bar{R}^{2/3}-1}{C^4-C}}\left(\frac{3\bar{R}}{4} - \frac{3\bar{R}^{1/3}}{8C^{4/3}}\right) - \frac{3\cdot\log\left(C^{2/3}\bar{R}^{1/3}+\sqrt{C^{4/3}\bar{R}^{2/3}-1}\right)}{8C^2\sqrt{C^4-C}}}{\displaystyle\frac{3C^2}{4} - \frac{3}{8C} - \frac{3\cdot\log\left(C^{3/2}+\sqrt{C^3-1}\right)}{8C^2\sqrt{C^4-C}}}$

Radius of Curvature Distribution in Oblate Spheroid
===

Mass function:

$\huge\displaystyle f_C(\bar{R}) = \frac{\sqrt{\frac{1-{C'}^{4/3}\cdot\bar{R}^{2/3}}{{C'}-{C'}^4}}}{-\frac{3{C'}^2}{4} + \frac{3}{8{C'}} + \frac{\frac{3\pi}{2} - 3\cdot\arcsin\left({C'}^{3/2}\right)}{8{C'}^2\sqrt{{C'}-{C'}^4}}}, \qquad C' = \frac{1}{C^2}$

Cumulative function:

$\large F_C(\bar{R}) = \frac{\displaystyle\sqrt{\frac{1-{C'}^{4/3}\bar{R}^{2/3}}{{C'}-{C'}^4}}\left(\frac{3\bar{R}}{4} - \frac{3\bar{R}^{1/3}}{8{C'}^{4/3}}\right) + \frac{3\cdot\arcsin\left({C'}^{2/3}\bar{R}^{1/3}\right)}{8{C'}^2\sqrt{{C'}-{C'}^4}}-\frac{3{C'}^2}{4} + \frac{3}{8{C'}} -  \frac{3\cdot\arcsin\left({C'}^{3/2}\right)}{8{C'}^2\sqrt{{C'}-{C'}^4}}}{\displaystyle -\frac{3{C'}^2}{4} + \frac{3}{8{C'}} +  \frac{\frac{3\pi}{2} - 3\cdot\arcsin\left({C'}^{3/2}\right)}{8{C'}^2\sqrt{{C'}-{C'}^4}}},\qquad \text{with } C' = \frac{1}{C^2}$

```python
from __future__ import print_function, division
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x, C):
    if C >= 1:
        if x < 1/C**2:
            return f(1/C**2, C)
        elif x > C**(5/2):
            return f(C**(5/2), C)
        numer = math.sqrt(abs(C**(4/3)*x**(2/3)-1))/math.sqrt(abs(C**4-C))
        denom = (3/4)*C**2 - (3/8)/C - 3*math.log(C**(3/2) + math.sqrt(abs(C**3-1)))/(8*C**(2)*math.sqrt(abs(C**4-C)))
    if C <= 1:
        if x > 1/C**2:
            return f(1/C**2, C)
        elif x < C**(5/2):
            return f(C**(5/2), C)
        numer = math.sqrt(abs(C**(4/3)*x**(2/3)-1))/math.sqrt(abs(C**4-C))
        denom = -(3/4)*C**(2) + (3/8)*C**(-1) + (3*np.pi/2 - 3*np.arcsin(C**(3/2)))/(8*C**(2)*math.sqrt(abs(C**4-C)))
    return numer/denom

def F(x, C):
    if C >= 1:
        if x < 1/C**2:
            return F(1/C**2, C)
        if x > C**(5/2):
            return F(C**(5/2), C)
        numer = math.sqrt(abs((C**(4/3)*x**(2/3)-1)/(C**4-C)))*((3/4)*x - (3/8)*(x**(1/3)/C**(4/3))) - (3*np.log(C**(2/3)*x**(1/3) + math.sqrt(abs(C**(4/3)*x**(2/3)-1))))/(8*C**(2)*math.sqrt(abs(C**4-C)))
        denom = (3/4)*C**2 - (3/8)/C - 3*np.log(C**(3/2) + math.sqrt(abs(C**3-1)))/(8*C**(2)*math.sqrt(abs(C**4-C)))
    else:
        if x < C**(5/2):
            return F(C**(5/2), C)
        if x > 1/C**2:
            return F(1/C**2, C)
        numer = math.sqrt(abs((C**(4/3)*x**(2/3)-1)/(C**4-C)))*((3/4)*x - (3/8)*(x**(1/3)/C**(4/3))) + (3*np.arcsin(C**(2/3)*x**(1/3)))/(8*C**(2)*math.sqrt(abs(C**4-C)))-(3/4)*C**(2) + (3/8)*C**(-1) - (3*np.arcsin(C**(3/2)))/(8*C**(2)*math.sqrt(abs(C**4-C)))
        denom = -(3/4)*C**(2) + (3/8)*C**(-1) + (3*np.pi/2 - 3*np.arcsin(C**(3/2)))/(8*C**(2)*math.sqrt(abs(C**4-C)))
    result = numer/denom
    return result
```

```python
num_samples = 100
Cs = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 1/1.5**2, 1/2**2, 1/2.5**2, 1/3**2, 1/3.5**2, 1/4**2]
for C in Cs:
    if C not in [1/(1.5**2), 3]:
        continue
    if C >= 1:
        x_min = 1/C**2
        x_max = C**(5/2)
    else:
        x_min = C**(5/2)
        x_max = 1/C**2
    x_list = np.linspace(x_min, x_max, num_samples+1)
    yf_list = []
    yF_list = []
    print(f'{F(x_min, C):.6f} - {F(x_max, C):.6f}')
    for x in x_list:
        yf_list.append(f(x,C))
        yF_list.append(F(x,C))
    plt.figure(1, figsize=(10,10))
    plt.subplot(211)
    plt.plot(x_list, yf_list)
    plt.xlabel('Normalized Radius of Curvature')
    plt.ylabel('Mass')
    plt.legend(['Prolate (3 uniaxially)', 'Oblate (1.5 biaxially)'], loc='upper right')

    plt.subplot(212)
    plt.plot(x_list, yF_list)
    plt.xlabel('Normalized Radius of Curvature')
    plt.ylabel('Cumulative')
    
    plt.legend(['Prolate (3 uniaxially)', 'Oblate (1.5 biaxially)'], loc='lower right')
```

```python
num_samples = 100
Cs = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 1/1.5**2, 1/2**2, 1/2.5**2, 1/3**2, 1/3.5**2, 1/4**2]
values = {C:[] for C in Cs}
for C in Cs:
    if C == 1:
        continue
#     if C not in [0.25, 2]:
#         continue
    if C >= 1:
        x_min = 1/C**2
        x_max = C**(5/2)
    else:
        x_min = C**(5/2)
        x_max = 1/C**2
    x_list = np.linspace(x_min, x_max, num_samples+1)
    yf_list = []
    yF_list = []
    
    for x in x_list:
        yf_list.append(f(x,C))
        yF_list.append(F(x,C))
        
#     plt.figure(1, figsize=(10,10))
#     plt.subplot(211)
#     plt.plot(x_list, yf_list)
#     plt.xlabel('Normalized Radius of Curvature')
#     plt.ylabel('Mass')

#     plt.subplot(212)
#     plt.plot(x_list, yF_list)
#     plt.xlabel('Normalized Radius of Curvature')
#     plt.ylabel('Cumulative')
    
    fig = plt.figure('Histogram for C={}'.format(C), figsize=(15,8))
    ax = fig.add_axes([0.15, 0.15, 0.8, 0.8])
    x_list = []
    x = round(x_min)
    if C != 1:
        n_bins = round(x_max)-round(x_min)+1
        offset = abs(x_max-x_min)/(2*n_bins)
        midx_list = np.linspace(x_min+offset, x_max-offset, n_bins)
        midx_idx = 0
    bins = [x-0.5]
    weights = []
    while x <= round(x_max):
        x_list.append(x)
        bins.append(x+0.5)
        if C==1:
            weights.append(1)
        else:
            mid_x = midx_list[midx_idx]
            midx_idx += 1
            mid_x = x
            offset = 0.5
            weights.append(F(mid_x+offset, C)-F(mid_x-offset, C))
        x += 1
    if C==1:
        x_list = [0,1,2,3,4]
        bins = [-0.5, 0.5, 1.5, 2.5, 3.5, 4.5]
        weights = [0,1,0,0,0]
        # ax.bar(x_list, weights, width=0.8, align='center')
        res = ax.hist(x_list, bins=bins, weights=weights, rwidth=0.8)
#     elif C==1.5:
#         x_list.append(4)
#         bins.append(4.5)
#         weights.append(0)
#         res = ax.hist(x_list, bins=bins, weights=weights, rwidth=0.8)
    else:
        # ax.bar(x_list, weights, width=0.8, align='center')
        res = ax.hist(x_list, bins=bins, weights=weights, rwidth=0.8)
    values[C] = weights
    # ax.set_title(r'$C={:g}$'.format(C))
    ax.set_xlabel(r'Relative Radius of Curvature, $R_C(t)$')
    ax.set_ylabel('Distribution')
    labels = list(range(n_bins))
    threshold = 24
    jump = 2
    labels[0] = '0.0-0.5'
    while len(labels) > threshold:
        for i in range(jump//2, len(labels), jump):
            labels[i] = ''
        threshold += 8*jump
        jump *= 2
    ax.set_xticks([i for i,label in enumerate(labels) if label != ''])
    ax.set_xticklabels([label for label in labels if label != ''], rotation=0, ha='center')
    if C > 1:
        legend_name = f'C={C} (prolate, uniaxial)'
        loc = 'upper left'
    elif C < 1:
        legend_name = f'C={1/math.sqrt(C):g} (oblate, biaxial)'
        loc = 'upper right'
    else:
        legend_name = f'C=1 (sphere)'
        loc = 'upper left'
    ax.legend([legend_name], loc=loc)

    plt.savefig('../C{:g}-hist.png'.format(C), bbox_inches='tight')
# plt.show()
```

```python
import pandas as pd
df = None
for C, weights in values.items():
    weights = weights[:]
    weights[0:0] = [f'C={C:g} (prolate)' if C >= 1 else f'C={1/C**0.5:g} (oblate)']
    cur_df = pd.DataFrame({f'{C}': weights})
    if df is None:
        df = cur_df
    else:
        df = pd.concat([df, cur_df], axis=1, ignore_index=True)
```
